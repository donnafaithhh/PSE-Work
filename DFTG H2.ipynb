```python
pathfile = r'fin_files/H2_voids.pkl'
with open(pathfile, 'rb') as f:
    data = pickle.load(f)
```

```python
plt.figure(figsize=(15,4))
plt.plot(data['holes'].iloc[0:2610-110])
plt.ylabel('Number of voids')
plt.xlabel('Time')
plt.xticks(range(0, 2610, 261))
plt.title('Number of H2 Voids in the PSE')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(25,4))
plt.plot(data['persistence'].iloc[0:2610-110])
plt.ylabel('Number of voids')
plt.xlabel('Time')
plt.xticks(range(0, 2610, 261))
plt.title('H2 Voids in the PSE')
plt.grid()
plt.show()
```

```python
data = data.iloc[0:2610-110]
data.columns
```

```python
len(data)
```

```python
year = '(2013-2023)'
max_len = len(data)
fig, axes = plt.subplots(3, 1, figsize=(16, 7), sharex=True)

axes[0].plot(data['holes'], label=year, color='b')
axes[0].set_title(f'H2 Voids for PSE {year}')
plt.xticks(range(0, max_len, 261))
# axes[0].set_ylim(0, 6)
# axes[0].set_xlim(0, 1635)
axes[0].grid()

axes[1].plot(data['death time'], label='death', color='r')
axes[1].plot(data['birth time'], label='birth', color='g')
axes[1].set_title(f'Comparison of the birth and death time of H2 Voids {year}')
plt.xticks(range(0, max_len, 261))
# axes[1].set_xlim(0, 1635)
axes[1].grid()
axes[1].legend()

x = range(len(data.index))
y = data['persistence']
axes[2].bar(x, y, label='persistence', color='b')
axes[2].set_title(f'Persistence of H2 Voids {year}')
plt.xticks(range(0, max_len, 261))
# axes[2].set_xlim(0, 1635)
axes[2].grid()

plt.xlabel("Window")
plt.tight_layout()
plt.show()
```
